## CLASSES

```python
################# http://www.pygame.org/wiki/2DVectorClass ##################
class Vec2d(object):
    """2d vector class, supports vector and scalar operators,
       and also provides a bunch of high level functions
       """
    __slots__ = ['x', 'y']

    def __init__(self, x_or_pair, y = None):
        if y == None:
            self.x = x_or_pair[0]
            self.y = x_or_pair[1]
        else:
            self.x = x_or_pair
            self.y = y

    # Addition
    def __add__(self, other):
        if isinstance(other, Vec2d):
            return Vec2d(self.x + other.x, self.y + other.y)
        elif hasattr(other, "__getitem__"):
            return Vec2d(self.x + other[0], self.y + other[1])
        else:
            return Vec2d(self.x + other, self.y + other)

    # Subtraction
    def __sub__(self, other):
        if isinstance(other, Vec2d):
            return Vec2d(self.x - other.x, self.y - other.y)
        elif (hasattr(other, "__getitem__")):
            return Vec2d(self.x - other[0], self.y - other[1])
        else:
            return Vec2d(self.x - other, self.y - other)

    # Vector length
    def get_length(self):
        return math.sqrt(self.x**2 + self.y**2)

    # rotate vector
    def rotated(self, angle):
        cos = math.cos(angle)
        sin = math.sin(angle)
        x = self.x*cos - self.y*sin
        y = self.x*sin + self.y*cos
        return Vec2d(x, y)
```

```python
# Compute path length
## start - Add your code here
def BM_2d_df(BM_2d_df = pd.DataFrame):
    dis_BM = np.array([np.sqrt(np.sum(np.square(BM_2d_df.loc[i-1] - BM_2d_df.iloc[i]))) for i in range(1,BM_2d_df.shape[0])])
    return np.cumsum(dis_BM)

pl_BM_3 = BM_2d_df(BM_2d_df_3)
pl_BM_6 = BM_2d_df(BM_2d_df_6)
pl_BM_9 = BM_2d_df(BM_2d_df_6)
## end - Add your code here
```

## Actividad 3: Turning-angle Distribution - (Dist. origen vs Dist. observada)

* Generar CRWs con dos exponentes diferentes
* Guardar trayectorias en Pandas Data Frame
* Obtener Turning-angle distribution
* Guardar metricas en **Pandas** Data Frame
* Comparar en gráfica distribución origen vs distribución observada (Histograma)
* Visualizar con **plotly**

```python
# Load existing trajectories
# CRW speed = 6,
# wrapcauchy [c = 0.6]
CRW_2d_df_6 = pd.read_csv('trajectories/crw_6_6.csv')

# Load existing trajectories
# CRW speed = 6,
# wrapcauchy [c = 0.9]
CRW_2d_df_9 = pd.read_csv('trajectories/crw_6_9.csv')
```

```python
# Init variables
CRW_exponent6 = 0.6
CRW_exponent9 = 0.9

# Resolution
resolution6 = CRW_2d_df_6.shape[0]
resolution9 = CRW_2d_df_9.shape[0]


aux_domain6 = np.linspace(0, 2*np.pi, resolution6)
aux_domain9 = np.linspace(0, 2*np.pi, resolution9)

wrapcauchy_pdf6 = np.array([wrapcauchy.pdf(i, CRW_exponent6) for i in aux_domain6])
wrapcauchy_pdf9 = np.array([wrapcauchy.pdf(j, CRW_exponent9) for j in aux_domain9])

maxD6 = np.max(aux_domain6)
for k in range(len(aux_domain6)):
    if (aux_domain6[k]>maxD6/2):
        aux_domain6[k] = aux_domain6[k]-maxD6

maxD9 = np.max(aux_domain9)
for l in range(len(aux_domain9)):
    if (aux_domain9[l]>maxD9/2):
        aux_domain9[l] = aux_domain9[l]-maxD9
```

```python
def turning_angles(trajectory, index):
    sideA = distance.euclidean(trajectory.iloc[index-1], trajectory.iloc[index])
    sideB = distance.euclidean(trajectory.iloc[index], trajectory.iloc[index+1])
    sideC = distance.euclidean(trajectory.iloc[index+1], trajectory.iloc[index-1])
    return(math.cos((math.pow(sideA, 2)+math.pow(sideC, 2)-math.pow(sideB, 2))/2*sideA*sideC))
```

```python
def search_change(trajectory, index):
    m1 = (trajectory.y_pos[index] - trajectory.y_pos[index-1]) / (trajectory.x_pos[index] - trajectory.x_pos[index-1])
    #print(m1)
    m2 = (trajectory.y_pos[index+1] - trajectory.y_pos[index]) / (trajectory.x_pos[index+1] - trajectory.x_pos[index])
    #print(m2)
    if(m1 == m2):
        return True
    else:
        return False
```

```python
# aux to store turning angles
aux_ta_CRW_2d_df_6 = np.empty(shape=(0))
CRW_2d_df_list6 = []
# Iterate over trajectory compute turning angles
for index, row in CRW_2d_df_6[1:-1].iterrows():
    ## start - Add your code here
    CRW_2d_df_list6.append(turning_angles(CRW_2d_df_6, index))
    ## end - Add your code here
aux_ta_CRW_2d_df_6_ = np.asarray(CRW_2d_df_list6)

for i in range(len(aux_ta_CRW_2d_df_6_)):
    aux_ta_CRW_2d_df_6_[i] = aux_ta_CRW_2d_df_6_[i]+1

maxR6 = np.max(aux_ta_CRW_2d_df_6_)
for j in range(len(aux_ta_CRW_2d_df_6_)):
    if (aux_ta_CRW_2d_df_6_[j]>maxR6/2):
        aux_ta_CRW_2d_df_6_[j] = aux_ta_CRW_2d_df_6_[j]-maxR6


# aux to store turning angles
aux_ta_CRW_2d_df_9 = np.empty(shape=(0))
CRW_2d_df_list9 = []
# Iterate over trajectory compute turning angles
for index, row in CRW_2d_df_9[1:-1].iterrows():
    ## start - Add your code here
    CRW_2d_df_list9.append(turning_angles(CRW_2d_df_9, index))
    ## end - Add your code here'''
aux_ta_CRW_2d_df_9_ = np.asarray(CRW_2d_df_list9)

for k in range(len(aux_ta_CRW_2d_df_9_)):
    aux_ta_CRW_2d_df_9_[k] = aux_ta_CRW_2d_df_9_[k]+1

maxR9 = np.max(aux_ta_CRW_2d_df_9_)
for l in range(len(aux_ta_CRW_2d_df_9_)):
    if (aux_ta_CRW_2d_df_9_[l]>maxR9/2):
        aux_ta_CRW_2d_df_9_[l] = aux_ta_CRW_2d_df_9_[l]-maxR9
# Save to pandas DF
## start - Add your code here
TD_CRW_2d_df_6 = pd.DataFrame(aux_ta_CRW_2d_df_6_)
TD_CRW_2d_df_9 = pd.DataFrame(aux_ta_CRW_2d_df_9_)
## end - Add your code here

# Write to csv
## start - Add your code here
TD_CRW_2d_df_6.to_csv("G:/Mi unidad/MCPA/2º_Semestre/Tópico_de_la_Industria_I/Semana3_03-09-22/metricsTD_CRW6.csv")
TD_CRW_2d_df_9.to_csv("G:/Mi unidad/MCPA/2º_Semestre/Tópico_de_la_Industria_I/Semana3_03-09-22/metricsRD_CRW9.csv")
## end - Add your code here
```
